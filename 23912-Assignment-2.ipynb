**Importing required libraries**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

**Creating a dataframe 1**

```python
df1 = pd.read_csv(r'C:\Users\Msc 1\Desktop\Sumu\MDSC-106 P\churn1.csv')
df1
```

**Creating a dataframe 2**

```python
df2 = pd.read_csv(r'C:\Users\Msc 1\Desktop\Sumu\MDSC-106 P\churn2.csv')
df2
```

**Merging two dataframes with CustomerID as foriegn key**

```python
df = df1.merge(df2,on='customerID')
df
```

```python
df.info()
```

**Converting the datatype of TotalCharges to numeric**

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.info()
```

**Checking if columns are merged properly**

```python
original_col = set(df1.columns).union(set(df2.columns))
merged_col = set(df)
if( original_col==merged_col ):
    print("All columns are merged properly")
else:
    print("Columns are not merged properly")
```

**Checking for Null values**

```python
df.isnull().sum()
```

**Plotting Pie charts for each feature**

```python
num_col = df.select_dtypes(include=['float', 'int']).columns
no_numcols = len(num_col)
obj_col = df.shape[1]
rows = (obj_col-no_numcols)//4
fig, axes = plt.subplots(rows, 4 , figsize=(15,12))
i,j = 0 , 0
for column_name in df:
    if(column_name != 'customerID' and  column_name not in num_col):
        category_counts = df[column_name].value_counts()
        ax = axes[i,j]
        ax.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
        ax.set_title(f'{column_name}')
        j+=1
        if(j > 3):
            j = 0
            i += 1
```

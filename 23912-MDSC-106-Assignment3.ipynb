**Setting the dtype of categorical features as object**

```python
features=['CreditCard', 'InternetBanking', 'FixedDepositAccount', 'Security', 'Level', 'HiddenScore', 'LoanOnCard']
for feature in features:
    df[feature]=df[feature].astype(object)

df.info()
```

**Percentage of missing values wrt to the feature:**

```python
for x in df:
    y=(df[x].isnull().sum())/df.shape[0]*100
    print('{}{}  {}%'.format(x,(22-len(x))*' ',y))
```

**Percentage of missing values in the dataframe:**

```python
import numpy as np
print('Percentage of missing values in the dataframe: {}%'.format(np.round((100*df.isnull().sum().sum())/(df.shape[0]*df.shape[1]),3)))
```

**Checking the possible values of each categorical feature**

```python
categorical_features=['CreditCard', 'InternetBanking', 'FixedDepositAccount', 'Security', 'Level', 'HiddenScore']
for x in df:
    if x in categorical_features:
        y=df[x].unique()
        print(x,':',(20-len(x))*' ',y)   
```

**Plotting distribution and histplots for int/float features, countplot for categorical_features**

```python
plt.figure(figsize=(20,15))
sns.color_palette("rocket")
i=1
for feature in df:
    if feature=='ID' or feature=='ZipCode':
        continue
    if df[feature].dtype=='int' or df[feature].dtype=='float':
        plt.subplot(4,3,i)
        sns.histplot(data=df, x=feature, kde=True, palette='rocket')
    else:
        plt.subplot(4,3,i)
        sns.countplot(data=df, x=feature, palette='icefire')
    i+=1
```

**Plotting a pairplot**

```python
sns.pairplot(data=df)
```

**Printing correlation between each set of features**

```python
df.corr()
```

**Plotting a heatmap**

```python
for feature in df:
    df[feature] = pd.to_numeric(df[feature], errors='coerce')

plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), annot=True, cmap="crest")
```

**Plotting Pie charts**

```python
plt.figure(figsize=(15,10))
i=1
for feature in categorical_features:
    feature_counts = df[feature].value_counts()

    plt.subplot(2,3,i)
    plt.pie(feature_counts, labels=feature_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'Pie Chart for {feature}')
    plt.axis('equal')
    
    i+=1
```
